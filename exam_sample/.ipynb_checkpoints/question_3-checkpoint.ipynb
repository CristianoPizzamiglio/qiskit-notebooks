```python
from qiskit import QuantumRegister, QuantumCircuit

inp_reg = QuantumRegister(2, name='inp')
ancilla = QuantumRegister(1, name='anc')
qc = QuantumCircuit(inp_reg, ancilla)
qc.draw()
```

```python
qc.h(inp_reg[0:2])
qc.x(ancilla)
qc.draw()
```

```python
qc_2 = QuantumCircuit(inp_reg, ancilla)
qc_2.h(inp_reg)
qc_2.x(ancilla)
qc_2.draw()
```
