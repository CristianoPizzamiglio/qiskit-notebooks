```python
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram, plot_bloch_vector
from math import sqrt, pi
```

```python
qc = QuantumCircuit(1) # Create a quantum circuit with one qubit.
initial_state = [0, 1] # Define inital_state as |1>
qc.initialize(initial_state, 0) # Apply initialization operation to the 0th quibit
qc.draw() # Let's view our circuit.
```

```python
backend = Aer.get_backend('statevector_simulator') # Tell Qiskit how to simulate our circuit to view the resulting state of our qubit.
```

```python
result = execute(qc, backend).result() # Perform the simulation using the 'statevector_simulator' and return the result.
```

```python
# From the result, we can get the final statevector:
out_state = result.get_statevector()
print(out_state) # Display the output state vector.
```

```python
# Now, let's measure our qubit as we would in a real quantum computer and see the result:
qc.measure_all()
qc.draw()
```

```python
counts = result.get_counts()
plot_histogram(counts)
# We can see that we (unsurprisingly) have a 100% chance of measuring |1>.
```

```python
# Now, let's put our qubit into a superposition state and see what happens and repeat the steps above:
initial_state = [1/sqrt(2), 1j/sqrt(2)] # Define state |q_0>
qc = QuantumCircuit(1)
qc.initialize(initial_state, 0)
state = execute(qc, backend).result().get_statevector()
print(state)
```

```python
results = execute(qc, backend).result().get_counts()
plot_histogram(results)
```

#### IMPORTANT:

##### Collapsing of the state of the qubit

It is a potent effect, and so one that must be used wisely.
For example, were we to constantly measure each of our qubits to keep track of their value at each point in a computation,
they would always simply be in a well-defined state of either 0 or 1. As such, they would be no different from classical bits
and our computation could be easily replaced by a classical computation. To achieve truly quantum computation we must allow
the qubits to explore more complex states. Measurements are therefore only used when we need to extract an output.
This means that we often place all the measurements at the end of our quantum circuit.

```python
# Let's demonstrate that:
qc = QuantumCircuit(1)
initial_state = [0.0 + 1j/sqrt(2), 1/sqrt(2)+ 0.0j]
qc.initialize(initial_state, 0)
qc.draw()
```

```python
state = execute(qc, backend).result().get_statevector()
print("Qubit State = " + str(state))
# The qubit is initialized in the state defined above.
```

```python
# But when we measure (observe) the qubit...
qc.measure_all()
qc.draw()
```

```python
# ...we can see that one of the amplitudes is always 0:
state = execute(qc, backend).result().get_statevector()
print("State of Measured Qubit = " + str(state))

# We can re-run this cell a few times to reinitialize the quibit and meaure it again.
# You will notice that either outcome is equally probable (50% of the times we can get 0 as amplitude).
# But the state of the qubit is never a superposition of |0> and |1>.
```

#### A Note about Quantum Simulators

We can see that writing down a qubits state requires keeping track of two complex numbers, but when using a real quantum computer we will only ever receive a yes-or-no (0 or 1) answer for each qubit. The output of a 10-qubit quantum computer will look like this:

0110111110

Just 10 bits, no superposition or complex amplitudes. When using a real quantum computer, we cannot see the states of our qubits mid-computation, as this would destroy them! This behaviour is not ideal for learning, so Qiskit provides different quantum simulators: The qasm_simulator behaves as if you are interacting with a real quantum computer, and will not allow you to use .get_statevector(). Alternatively, statevector_simulator, (which we have been using in this chapter) does allow peeking at the quantum states before measurement, as we have seen.
