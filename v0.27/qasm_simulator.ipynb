```python
from qiskit import QuantumCircuit, Aer, transpile, execute
from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_state_city
```

```python
# Create a Quantum Circuit acting on the q register
qc = QuantumCircuit(3, 3)

# Add a H gate on qubit 0
qc.h(0)

# Add a CX (CNOT) gate on control qubit 0 and target qubit 1
qc.cx(0, 1)

# Add a CX (CNOT) gate on control qubit 0 and target qubit 2
qc.cx(0, 2)

# Add barrier
qc.barrier(range(3))

# Map the quantum measurement to the classical bits
qc.measure([0, 1, 2], [0, 1, 2])

# Draw the circuit
qc.draw()
```

```python
# Choose the statevector simulator backend
backend = Aer.get_backend("qasm_simulator")
```

```python
# Transpile the circuit, not necessary
# qc_compiled = transpile(qc, backend)
```

```python
# Run the simulation
# job = backend.run(qc_compiled, shots=1000)
job = execute(qc, backend, shots=1000)
```

```python
# Get results from the simulation
results = job.result()
```

```python
counts = results.get_counts()
plot_histogram(counts)
```
