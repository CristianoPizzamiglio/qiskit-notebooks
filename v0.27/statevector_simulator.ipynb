```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_state_city
```

```python
# Create a Quantum Circuit acting on the q register
qc = QuantumCircuit(3, 3)

# Add a H gate on qubit 0
qc.h(0)

# Add a CX (CNOT) gate on control qubit 0 and target qubit 1
qc.cx(0, 1)

# Add a CX (CNOT) gate on control qubit 0 and target qubit 2
qc.cx(0, 2)

# Draw the circuit
qc.draw()
```

```python
# Choose the statevector simulator backend
backend = Aer.get_backend("statevector_simulator")
```

```python
# Run the simulation
# job = backend.run(qc)
job = execute(qc, backend)
```

```python
# Get results from the simulation
results = job.result()
```

```python
# Get the statevector for the quantum circuit
output_statevector = results.get_statevector()
print(output_statevector)
```

```python
counts = results.get_counts()
plot_histogram(counts)
```

```python
# Plot the real and imaginary components of the state density matrix rho
plot_state_city(output_statevector)
```
